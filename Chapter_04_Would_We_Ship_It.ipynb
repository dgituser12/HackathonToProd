## Hands-on Exercise 3

### Define Operational Metrics


Identify the metrics that should be monitored after
deployment.

Operational metrics help engineers detect issues before
customers report them.

```python
operational_metrics = {

    "Availability": "Percentage of successful requests",

    "Latency": "Average response time",

    "Error Rate": "Percentage of failed requests",

    "Accuracy": "Correct predictions",

    "Retry Count": "Average retries per request",

    "User Feedback": "Reported satisfaction",

    "Cost": "Average cost per request"

}

for metric, description in operational_metrics.items():

    print(f"{metric:15} {description}")
```

## Engineering Discussion


What engineers monitor depends on the goals of the system.

A fast service with poor accuracy is disappointing.

A highly accurate service that is unavailable is unusable.

Production AI requires balancing multiple quality
attributes rather than optimizing only one.

## Hands-on Exercise 6

### Plan a Deployment Strategy


Few applications are deployed to every user immediately.

Consider the advantages of deploying gradually.

```python
deployment_stages = [

    "Internal Testing",

    "Pilot Users",

    "Limited Rollout",

    "Regional Deployment",

    "Full Production"

]

for stage in deployment_stages:

    print(stage)
```

## Reflection


Incremental deployment reduces operational risk.

If problems appear during a pilot rollout, they affect a
small number of users rather than the entire customer base.

This strategy provides opportunities to learn before
expanding deployment.

## Hands-on Exercise 8

### Conduct a Production Readiness Review


Imagine you are the engineering review board responsible for
approving HelpDeskAI for production deployment.

Discuss each question before making your recommendation.

```python
review_questions = [

    "Is the model accurate enough?",

    "Can failures be detected quickly?",

    "Can the application recover from transient issues?",

    "Are incorrect predictions measurable?",

    "Can the engineering team operate this system at scale?",

    "Do we understand the operational costs?",

    "Is customer feedback incorporated into future improvements?"

]

for question in review_questions:

    print("-", question)
```

## Engineering Discussion


Production reviews rarely result in a simple 'yes' or 'no'.

Instead, engineering teams identify remaining risks,
prioritize improvements, and decide whether those risks are
acceptable for the intended users and business context.

Shipping software is ultimately a decision about confidence,
not certainty.
